## Pandas
- **you will represent your data using pandas**
- pandas can define, save, load and view DataFrame tables
- you can manipulate the tables

### (1) you can create a data table (DataFrame)

```python
## Create a simple dataset
data = {'Name': ["John", "Anna", "Peter", "Linda"],
        'Location' : ["New York", "Paris", "Berlin", "London"],
        'Age' : [24, 13, 53, 33]
       }

df = pd.DataFrame(data)
```

```python
df
```

### (2) you can filter the data that you want


```python
# you can filter the data that you want
df[df.Age > 30] # here : row-wise
```
